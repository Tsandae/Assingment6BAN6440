```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import Input
```

```python
# Load Data
df = pd.read_csv(r"C:\Users\User\Downloads\teleconnect.csv") 
```

```python
# Check the null value
df.isnull().sum()
```

```python
# Convert TotalCharges to numeric and handle errors
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df.dropna(inplace=True)
df.drop(['customerID'], axis=1, inplace=True)
```

```python
# Encode categorical features
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
# Features and Target
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
# Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Build ANN Model
def build_model(optimizer_name):
    model = Sequential()
    model.add(Input(shape=(X_train.shape[1],)))
    model.add(Dense(16,activation='relu') )
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer=optimizer_name, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Optimization Algorithms to Test
optimizers = ['sgd', 'adam', 'rmsprop']
results = {}
```

```python
# Train and Evaluate Models
for opt in optimizers:
    print(f"\nTraining with optimizer: {opt.upper()}")
    model = build_model(opt)
    history = model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=0, validation_split=0.2)
    y_pred = (model.predict(X_test) > 0.5).astype("int32")
    acc = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    results[opt] = {
        'accuracy': acc,
        'classification_report': report,
        'history': history.history,
        'model': model,
        'y_pred': y_pred
    }
```

```python
# Print Results
for opt in results:
    print(f"\n--- Optimizer: {opt.upper()} ---")
    print(f"Accuracy: {results[opt]['accuracy']:.4f}")
    print("Classification Report:")
    print(pd.DataFrame(results[opt]['classification_report']).transpose())
```

```python
# Visualizations
for opt in results:
    history = results[opt]['history']
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Train Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{opt.upper()} - Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
```

```python
plt.subplot(1, 2, 2)
plt.plot(history['loss'], label='Train Loss')
plt.plot(history['val_loss'], label='Validation Loss')
plt.title(f'{opt.upper()} - Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Confusion Matrix
cm = confusion_matrix(y_test, results[opt]['y_pred'])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title(f'{opt.upper()} - Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
